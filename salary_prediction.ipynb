```python
x = df.iloc[:,:4]
x.head()
#loc is location
#we got every column other than salary
```

```python
y=df.loc[:,['Salary']]
y=df.iloc[:,4:]
#Both are correct , follow whatever you want
#loc is select by lables , and iloc is select by value or position
y
```

```python
#divide data into dependent and independent , and then we further divide it 
#it into 2 more for train and test the data.
from sklearn.model_selection import train_test_split
#validatio = 0.2 = 20percent data 
#pip install scikit-learn
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=0)
#we have same value using random state , training on the same value.
#otherwise every other person had to tackle with diff random values and diff accuracy
#giving test size to 20 percent , just give this model will automatically 
#divide the data into test to 80 percent and test to 20 percent.
#x = independent and y is dependent variable
x_test
y_train
y_test
```

```python
#Now we will train the model 
#all algos are presnt inside the sklearnlibrary for ml 
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score #for accuracy
#classifier is for classification as it is presetn on both the clas and reg

#creating objects of the imported algorithms
lr1 = LogisticRegression()
lr = LinearRegression()
bt = DecisionTreeRegressor()
rf = RandomForestRegressor()


#fitting the data using objects and performing training on the data
lr.fit(x_train,y_train)
lr1.fit(x_train,y_train)
bt.fit(x_train,y_train)
rf.fit(x_train,y_train)



y_pred1 = lr.predict(x_test)
y_pred2 = lr1.predict(x_test)
y_pred3 = bt.predict(x_test)
y_pred4 = rf.predict(x_test)


#predicting the values using the predict function to see if the model is working
#correctly and how much it predicts
'''y_pred1 = lr.predict([[1.1,8,22,75]])
y_pred2 = lr1.predict([[1.1,8,22,75]])
y_pred3 = bt.predict([[1.1,8,22,75]])
y_pred4 = rf.predict([[1.1,8,22,75]])#2d because to get in row and column format
#writing the data and predicting the salary the guy should be offered'''



print("Original Salary is 39343")
print(f"Predict salary for Linear Regression {y_pred1} & accuracy is {r2_score(y_test,y_pred1)}")
print(f"Predict salary for Logistic Regression {y_pred2} & accuracy is {r2_score(y_test,y_pred2)}")
print(f"Predict salary for Decision Tree Regression {y_pred3} & accuracy is {r2_score(y_test,y_pred3)}")
print(f"Predict salary for Decision Tree Regression {y_pred4} & accuracy is {r2_score(y_test,y_pred4)}")
```
